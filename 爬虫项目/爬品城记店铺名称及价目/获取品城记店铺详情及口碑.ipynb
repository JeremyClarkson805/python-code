```python
import re 
import requests 
import time 
from time import sleep
from selenium import webdriver
from lxml import etree
path = r'./Driver_Notes/msedgedriver.exe'
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36 Edg/111.0.1661.62'}
url_tot = 'https://space.bilibili.com/429582883/video?tid=0&page='
```

获取oid

```python
def get_oid(url):
    response = requests.get(url=url,headers=headers).text
    FindOid = re.compile(r'aid":(.*?),"')
    oid = re.findall(FindOid,response)
#     print('oid=',oid[0])
    return str(oid[0])
```

获取每个视频下面的店铺信息及菜单

(r'"uname":"品城记本地版".*?"content":{"message":"(.*?)".*?"')

```python
def get_comment(oid):
    url_1 = 'https://api.bilibili.com/x/v2/reply/main?csrf=4e574def7a3fa2a6fc6161005963d260&mode=2&next=0&oid='
    url_3 = '&plat=1&seek_rpid=&type=1'
    url = url_1 + oid + url_3
    response = requests.get(url=url,headers = headers).text
    FindIntroduce = re.compile(r'"uname":"品城记本地版".*?"content":{"message":"(.*?)".*?"')
    introduce = re.findall(FindIntroduce,response)
    introduce = str(introduce)
    introduce = introduce.replace(r'\\n','\n') 
    print(introduce[0])
    print("\n",response)
```

```python
oid = get_oid('https://www.bilibili.com/video/BV16s4y1U72Y/?spm_id_from=333.999.0.0&vd_source=205259626d771a8c4b63b9b3a51918ba')
get_comment(oid)
```

获取一个网站中所有视频的url

```python
edge = webdriver.Edge(path)
url = 'https://space.bilibili.com/429582883/video'
edge.get(url)
time.sleep(3)
response = edge.page_source
print(type(response))
edge.close()
print(response)
```

```python
Findurl = re.compile(r'><a href="//(.*?)/" target="_blank" class="cover">')
url_1 = re.findall(Findurl,response)
del url_1[30:61]
print(url_1)
```

```python
def find_each_url(url):
    edge = webdriver.Edge(path)
    edge.get(url)
    time.sleep(3)
    response = edge.page_source
    print(type(response))
    edge.close()
    Findurl = re.compile(r'><a href="//(.*?)/" target="_blank" class="cover">')
    url_1 = re.findall(Findurl,response)
    del url_1[30:61]
    for url in url_1:
        get_comment(get_oid(url))
```

```python
import re 
import requests 
```

```python
url = 'https://www.bilibili.com/video/BV1Mx4y1779b/?spm_id_from=333.999.0.0&vd_source=205259626d771a8c4b63b9b3a51918ba'
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36 Edg/111.0.1661.62'}
```

```python
response = requests.get(url=url,headers=headers).text
print(response)
```

```python
Findaid = re.compile(r'aid"(.*?),"')
aid_list = re.findall(Findaid,response)
print(aid_list[0])
```

```python
Find_tittle = re.compile(r'</script> <title data-vue-meta="true">(.*?)_哔哩哔哩_bilibili</title>')
tittle = re.findall(Find_tittle,response)
print(tittle)
```

```python
Find_det = re.compile(r'name="keywords" content="(.*?)视频播放量')
det = re.findall(Find_det,response)
print(det)
```

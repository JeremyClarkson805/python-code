## Canny边缘检测

<font size = 3> Canny边缘检测算法被认为是边缘检测的最优算法<br>
Canny边缘检测的一般步骤:<br>
1.进行边缘检测前，一般用高斯滤波去除噪声<br>
2.计算梯度:对平滑后的图像采用Sobel算子计算梯度和方向</font> 

 ### 非极大值抑制(NMS算法):

<font size=3> 非极大值抑制(NMS算法):<br>
1.在获取梯度和方向后，遍历图像，去除所有不是边界的点<br>
2.实现方法:遍历像素点，判断当前像素点是否是周围像素点中具有相同方向梯度的最大值<br>
3.下图中，点A，B，C具有相同的方向，梯度方向垂直于边缘<br>
4.判断A是否为A，B，C中的局部最大值，如果是，则保留该点，否则，他被抑制(归零)</font>

<font size=3>比如说上图，C，A，B三点，A点为最大值，则说明A点在边界上,会保留A点的值，抑制其他点  </font>

### 滞后阈值:

### 原理:
<font size=3> 1.若某一像素位置的幅值超过高阈值，则该位置被保留<br>
2.若某一像素位置的幅值低于低阈值，则该位置被去除<br>
3.若某一像素位置的幅值处于高低阈值之间，则该像素仅仅在连接到一个高于高阈值像素时被保留</font>

<font size=3>最大值和最小值要靠自己的经验来调，来试<br>
cv2.Canny(img,MinVal,MaxVal)</font>

```python
import cv2
import numpy as np
img = cv2.imread('./1.aidaishu.jpg')
img = cv2.resize(img,(378,504))
img_1 = cv2.Canny(img,80,180)
cv2.imshow('Canny',img_1)
cv2.waitKey()
cv2.destroyAllWindows()
```

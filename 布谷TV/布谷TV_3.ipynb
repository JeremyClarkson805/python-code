```python
import requests
import re
from lxml import etree
import time
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.1722.34'}
```

```python
def getPage_Data(url_t):
    FindB = re.compile(r'.net/(.*?).html')
    Find_Download = re.compile(r'<p>(.*?)">magnet:?')
    response_1 = requests.get(url_t,headers = headers).text
    response_1 = str(response_1)
    # print(response_1)
    Links = re.findall(Find_Download,response_1)
#     print(response_1.text)
    tree_1 = etree.HTML(response_1)
    path_b = re.findall(FindB,url_t)
    path_b = str(path_b)
#     print(path_b)
    path_intro = '//*[@id=\"post-' + str(path_b) + '\"]/div/div[3]/div/p'
    path_intro = path_intro.replace("['","").replace("']",'')
#     print(path_intro)
    info_list = tree_1.xpath(path_intro)
#     print(info_list)
    f = open('./布谷TV电影.txt','a',encoding = 'utf-8')
    for inf_1 in info_list:
        inf_t = inf_1.xpath('./text()')
        for inf in inf_t:
            print(inf)
            inf = str(inf)
            inf = inf.replace('复制磁力链接后打开 迅雷、utorrent、Bitcomet等进行下载！','')
            inf = inf.replace('点击或者复制磁力链接后打开迅雷、utorrent、Bitcomet等进行下载！','')
            f.write(inf)
            f.write('\n')
    print('\n',"下载链接:")
    for link in Links:
        link = str(link)
        link = link.replace('<br><a href="','\n')
        print(link)
        f.write(link)
        f.write('\n')
    f.write('\n\n')
    f.close()
```

```python
def main():
    base_url_1 = 'https://www.bugutv.net/4kuhd/page/'
    for i in range(1,2):
        url_1 = base_url_1 + str(i)
        url_1 = 'https://www.bugutv.net/4kuhd'
        response = requests.get(url_1,headers)
        tree = etree.HTML(response.text)#实例化xpath要把response换成text再丢进去
        div_list = tree.xpath('//*[@id="main"]/div[2]/div[1]/div/div/div[1]/div')# //方式是定位整个页面文档中所有符合的元素
        href_list = []
        for div in div_list:
            name = div.xpath('./article/div[2]/header/h2/a/text()')
            href = div.xpath('./article/div[2]/header/h2/a/@href')
            href_list.append(href)
        for href in href_list:
            href = str(href)
            href = href.replace("['",'').replace("']",'')
            getPage_Data(href)
```

```python
if __name__ == "__main__":
    start_time = time.time()
    main()
    end_time = time.time()
    print('Spend %f Second'%(end_time-start_time))
```

```python
def get_response(url_1):               #获得一整个页面的源码
    response_1 = requests.get(url = url_1,headers = headers)
#     print(response_1.text)
    url_2 = 'https://www.bugutv.net/4kmovie'
    response_2 = requests.get(url = url_2,headers = headers).text
    tree_2 = etree.HTML(response_2)
    divs = tree_2.xpath('//*[@id="main"]/div[2]/div[1]/div/div/div[1]/div')
    href_list = []
    for div in divs:
        href = div.xpath('./article/div[2]/header/h2/a/@href')   
        href = str(href)
        href = href.replace("['",'').replace("']",'')
        print(href)
        href_list.append(href)
    pool_1 = Pool(20)
    pool_1.map(getPage_Data,href_list)
    print('done')
```

```python
def getPage_Data(url_t):
    FindB = re.compile(r'.net/(.*?).html')
    Find_Download = re.compile(r'<p>(.*?)">magnet:?')
    response_1 = requests.get(url_t,headers = headers).text
    response_1 = str(response_1)
    # print(response_1)
    Links = re.findall(Find_Download,response_1)
#     print(response_1.text)
    tree_1 = etree.HTML(response_1)
    path_b = re.findall(FindB,url_t)
    path_b = str(path_b)
#     print(path_b)
    path_intro = '//*[@id=\"post-' + str(path_b) + '\"]/div/div[3]/div/p'
    path_intro = path_intro.replace("['","").replace("']",'')
#     print(path_intro)
    info_list = tree_1.xpath(path_intro)
#     print(info_list)
    f = open('./布谷TV电影_f.txt','a',encoding = 'utf-8')
    for inf_1 in info_list:
        inf_t = inf_1.xpath('./text()')
        for inf in inf_t:
            print(inf)
            inf = str(inf)
            inf = inf.replace('复制磁力链接后打开 迅雷、utorrent、Bitcomet等进行下载！','')
            inf = inf.replace('点击或者复制磁力链接后打开迅雷、utorrent、Bitcomet等进行下载！','')
            f.write(inf)
            f.write('\n')
    print('\n',"下载链接:")
    for link in Links:
        link = str(link)
        link = link.replace('<br><a href="','\n')
        print(link)
        f.write(link)
        f.write('\n')
    f.write('------------------------------------------------------------------------------------')
    f.write('\n\n')
    f.close()
```

```python
def main():
    start_time_2 = time.time()
    base_url_1 = 'https://www.bugutv.net/4kuhd/page/'
    base_url_list = []
    for i in range(1,2):
        url_1 = base_url_1 + str(i)
        print(url_1)
        base_url_list.append(url_1)
    pool_2 = Pool(10)
    pool_2.map(get_response,base_url_list)
    print("爬取完成！！！")
    end_time_2 = time.time()
    print('Spend %f Second'%(end_time_2-start_time_2))
```
